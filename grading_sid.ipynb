```python
# Custom aggregation function to count the number of trades
def count_trades(x):
    return x.nunique()
```

```python
df1 = pd.pivot_table(df, values = ['total_seats', 'Admission_2023', 'female_student', 

                'dst_enrollment','sc_st_pwd', 'trade', 'trade_type'],
    aggfunc={
        'total_seats': 'sum',
        'Admission_2023': 'sum',
        'female_student': 'sum',
        'dst_enrollment': 'sum',
        'sc_st_pwd': 'sum',
    },
    index=pd.Grouper(key="misiticode"),
).reset_index()
```

```python
df1
```

```python
# Separate pivot table to count the number of trades based on their trade_type
trade_count_pivot = df.pivot_table(
    index='misiticode',
    columns='trade_type',
    values='trade',
    aggfunc='count'
).reset_index()
```

```python
trade_count_pivot.shape
```

```python
# Merging the two pivot tables
df2 = pd.merge(df1, trade_count_pivot, on='misiticode', how='left')
```

```python
df2.head()
```

```python
# Select the columns to merge from df2 into df1
columns_to_merge = ['misiticode', 'ITI_Name', 'date_of_establishment',
       'iti_category', 'state', 'districtname']
```

```python
df_merged = pd.merge(df2, df[columns_to_merge].drop_duplicates(), on='misiticode', how='left')
```

```python
df_merged.columns
```

```python
df_merged["Engineering_Count"] = 0
 
# df.loc[(df["trade_type"] == "Engineering") & (df["trade"] >= 1), "Engineering"] = 1
df_merged.loc[(df_merged["Engineering"] >= 1), "Engineering_Count"] = 1
```

```python
df_merged.rename(columns={'Non Engineering' : 'Non_Eng'}, inplace = True)
```

```python
df_merged["Non_Eng_Count"] = 0
 
# df.loc[(df["trade_type"] == "Engineering") & (df["trade"] >= 1), "Engineering"] = 1
df_merged.loc[(df_merged["Non_Eng"] >= 1), "Non_Eng_Count"] = 1
```
