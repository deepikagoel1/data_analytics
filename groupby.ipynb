```python
df1 = df.groupby('Sector', dropna = True).agg({'Name of the industries': lambda x: list(x)}).reset_index()
```

```python
df1.rename(columns={'Name of the industries': 'Industry_Name'}, inplace = True)
```

```python
df1
```

```python
from itertools import chain
import numpy as np

```

```python
# Flatten the Industry_Name list and repeat Sector values accordingly
industry_flattened = list(chain.from_iterable(df1.Industry_Name))
sector_repeated = np.repeat(df1.Sector.values, [len(x) for x in df1.Industry_Name])

# Create the new DataFrame
df2 = pd.DataFrame({
    'Sector': sector_repeated,
    'Industry_Name': industry_flattened
})

print(df2)
```

```python
df2.to_excel("Industry_Sector.xlsx", index = 0)
```
